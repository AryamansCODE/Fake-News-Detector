## Pre processing TEXT 
### creating a function to convert the text in lowercase , remove extra spaces and speacial chars, urls and links

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+',b'',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
# this part can take a few seconds due to large dataset

data['text'] = data['text'].apply(wordopt)
```

Defining dependant and independant variable x and y

```python
x = data['text']
y = data['class']
```

## Training the model
#### splitting dataset into training and testing set

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

## Extracting features from the text

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert text data into numerical format using TF-IDF to capture the importance of words in documents.
# This enables machine learning models to process and analyze text data effectively.
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print (classification_report(y_test, pred_lr))
```

# Testing the Model

```python
def output_lable(n):
    if n==0:
        return "Fake News"
    elif n==1:
        return "Correct News"
    
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test['text'] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
#     pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)
    
    return print("\n\nLR Predicition: {} \nDT Prediction: {} \nRFC Prediction:{}".format(output_lable(pred_LR[0]),
                                                                                                             output_lable(pred_DT[0]),
                                                                                                             output_lable(pred_RF[0])))
```
